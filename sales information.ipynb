```python
the_city=sales["City"]
the_sales=sales["Sales"]
the_date=sales["Date"]
```

```python
plt.title("date per sales")
plt.xlabel("Sales")
plt.ylabel("Date")
plt.plot(the_sales,the_date,color='y')
plt.show()
```

```python
plt.title("Sales per city")
plt.xlabel("city")
plt.ylabel("sales")
plt.bar(the_city,the_sales,width=0.5,color='r')
plt.show()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2)
ax[0].set_title("sales per date")
ax[0].plot(the_sales,the_date,color='y')
ax[1].set_title("sales per city")
ax[1].bar(the_city,the_sales,width=0.5,color='r')
```
