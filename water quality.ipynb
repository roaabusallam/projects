# Scan and clean data

```python
water_q.sample(10)
```

```python
water_q.dropna(subset=['Date'],inplace=True)
```

```python
water_q.isnull().sum()
```

```python
#change string to date 
water_q['Date']=pd.to_datetime(water_q['Date'])
```

```python
water_q=water_q.sort_values(by='Date', ascending=True)
water_q
```

```python
#fill NaN values with mean

water_q=water_q.fillna({'Salinity (ppt)':0.7,
                        'DissolvedOxygen (mg/L)':6.6,
                        'pH':7.2,'WaterDepth (m)':0.8,
                        'SecchiDepth (m)':0.5,
                        'WaterDepth (m)':0.8,
                        'WaterTemp (C)':18.1})
water_q
```

```python
water_q.isnull().sum()
```

# Data plot

```python
water_q
```

```python
# Water depth over the past 30 years in bar plot
plt.figure(figsize=(15,10))
sns.barplot(x=water_q['Date'].dt.year, y=water_q['WaterDepth (m)'],palette='spring')
plt.xlabel('years')
plt.ylabel('WaterDepth (m)')
plt.title('Water depth over the past 30 years')
plt.show()
```

```python
#Line plot
plt.figure(figsize=(15,10))
plt.plot(water_q['Date'].dt.year,water_q['Salinity (ppt)'],color='r',linestyle='--',marker='o',label='Salinity')
plt.plot(water_q['Date'].dt.year,water_q['DissolvedOxygen (mg/L)'],color='y',linestyle='--',marker='o',label='DissolvedOxygen')
plt.plot(water_q['Date'].dt.year,water_q['pH'],color='b',linestyle='--',marker='o',label='ph')
plt.title('Water Quality')
plt.xlabel('years')
plt.ylabel('scale')
plt.legend()
plt.grid()
plt.show()
```
